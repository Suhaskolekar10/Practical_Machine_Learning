```python
playground=pd.read_csv('train.csv')
dum_pground = pd.get_dummies(playground,drop_first=True)
X=dum_pground.drop('smoking',axis=1)
y=dum_pground['smoking']    
```

```python
from lightgbm import LGBMRegressor,LGBMClassifier
lgbm=LGBMRegressor(random_state=23)
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
params={'max_depth':[2,3,4,5,6,None],'learning_rate':np.linspace(0.0001,0.999,10) ,'n_estimators':[50,100,150]}
gcv_lgbm=GridSearchCV(lgbm,param_grid=params,cv=kfold,n_jobs=5)
gcv_lgbm.fit(X,y)
print(gcv_lgbm.best_params_)
print(gcv_lgbm.best_score_)
```

```python
from xgboost import XGBRegressor
xgbm=XGBRegressor(random_state=23)
params={'max_depth':[2,3,4,5,6,None],'learning_rate':np.linspace(0.0001,0.999,10) ,'n_estimators':[50,100,150]}
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
gcv_xgbm=GridSearchCV(xgbm,param_grid=params,cv=kfold,n_jobs=5)
gcv_xgbm.fit(X,y)
print(gcv_xgbm.best_params_)
print(gcv_xgbm.best_score_)
```

```python
from catboost import CatBoostRegressor
cgbm=CatBoostRegressor(random_state=23)
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
params={'max_depth':[2,3,4,5,6,None],'learning_rate':np.linspace(0.0001,0.999,10) ,'n_estimators':[50,100,150]}
gcv_cgbm=GridSearchCV(cgbm,param_grid=params,cv=kfold,n_jobs=5)
gcv_cgbm.fit(X,y)
print(gcv_cgbm.best_params_)
print(gcv_cgbm.best_score_)
```
