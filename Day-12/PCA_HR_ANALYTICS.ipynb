```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np  
import seaborn as sns 
import matplotlib.pyplot as plt 
prcomp=PCA()
```

```python
hr=pd.read_csv('HR_comma_sep.csv')
hr=pd.get_dummies(hr,drop_first=True)
```

```python
X=hr.drop('left',axis=1)
y=hr['left']
```

```python
#Stratified KFold is used for classification
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV,KFold,StratifiedKFold
from sklearn.pipeline import Pipeline
prcomp=PCA()
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
svm=SVC(kernel='linear')
pipe_svc=Pipeline([('PC',prcomp),('SVM',svm)])        
params={'PC__n_components': [2,3,4,5,6,7,8,9]}
gcv = GridSearchCV(pipe_svc, param_grid=params,cv=kfold,verbose=3)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_selector,make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV,KFold,StratifiedKFold
from sklearn.pipeline import Pipeline
hr=pd.read_csv('HR_comma_sep.csv')
ohc=OneHotEncoder(sparse_output=False,drop='first')
ct_enc=make_column_transformer((ohc,make_column_selector(dtype_include=object)),("passthrough",make_column_selector(dtype_include=['int64','float64'])),verbose_feature_names_out=False).set_output(transform="pandas")

#Column Transformer is better in the long run

scaler=StandardScaler()
#prcomp=PCA()                                                                                                                                   prcomp=PCA()
#kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
svm=SVC(kernel='linear')
pipe_svc=Pipeline([('TRNF',ct_enc),('SCL',scaler),('PC',prcomp),('SVM',svm)]) #Sequence in pipeline is important
params={'PC__n_components': [2,3,4,5,6,7,8,9]}
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
gcv = GridSearchCV(pipe_svc, param_grid=params,cv=kfold,verbose=3)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)                                                                                                                                                                                                                                                                                                                                                            
```

```python
#Adding C Paramater
hr=pd.read_csv('HR_comma_sep.csv')
ohc=OneHotEncoder(sparse_output=False,drop='first',max_categories=10)
ct_enc=make_column_transformer((ohc,make_column_selector(dtype_include=object)),("passthrough",make_column_selector(dtype_include=['int64','float64'])),verbose_feature_names_out=False).set_output(transform="pandas")

#Column Transformer is better in the long run

scaler=StandardScaler()
#prcomp=PCA()                                                                                                                                   prcomp=PCA()
#kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
svm=SVC(kernel='linear')
pipe_svc=Pipeline([('TRNF',ct_enc),('SCL',scaler),('PC',prcomp),('SVM',svm)]) #Sequence in pipeline is important
params={'PC__n_components': [2,3,4,5,6,7,8,9],'SVM__C':np.linspace(0.001,5,10)}
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
gcv = GridSearchCV(pipe_svc, param_grid=params,cv=kfold,verbose=3)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)                                                                                                                                                                                                                                                                                                                                                            
```
