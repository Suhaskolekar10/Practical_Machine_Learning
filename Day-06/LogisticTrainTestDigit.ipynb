```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold,cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss,accuracy_score
from sklearn.linear_model import ElasticNet
```

# 1.Use Logistic_Regression


```python
digitr=pd.read_csv('train.csv')
X=digitr.drop('label',axis=1)
y=digitr['label']
```

```python
digitr
```

```python
#logloss
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)

lr=LogisticRegression()
params={'penalty':["l1","l2",None],
        'multi_class':['multinomial']}
#Default scoring-acc score
gcv=GridSearchCV(lr,param_grid=params,verbose=2,cv=kfold,scoring="neg_log_loss",n_jobs=-1)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#Inferencing
bm_logistic=gcv.best_estimator_
test=pd.read_csv(r"test.csv")
y_pred=bm_logistic.predict(test)
submit_lg=pd.DataFrame({'ImageId':np.arange(1,28001),'label':y_pred})
submit_lg.to_csv(r"/home/dai/Desktop/Practical Machine Learning/Day-06/submit.csv",index=False)
```

# KNN


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import ElasticNet
from sklearn.metrics import log_loss,accuracy_score
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
scalar=MinMaxScaler()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=23,stratify=y)
X_scl_trn=scalar.fit_transform(X_train)
X_scl_tst=scalar.transform(X_test)
```

```python
#With Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
knn=KNeighborsClassifier() 
scalar=MinMaxScaler()
pipe=Pipeline([("SCL",scalar),("KNN",knn)])          
params={'KNN__n_neighbors':np.arange(1,6)}

#here we are using values of KNN transformer
#log_loss
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#Inferencing
bm_logistic=gcv.best_estimator_
test=pd.read_csv(r"test.csv")
y_pred=bm_logistic.predict(test)
submit_lg=pd.DataFrame({'ImageId':np.arange(1,28001),'label':y_pred})
submit_lg.to_csv(r"/home/dai/Desktop/Practical Machine Learning/Day-06/submitKNN.csv",index=False)
```
