```python
###-----------------
### Import Libraries
###-----------------

import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold,GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis,QuadraticDiscriminantAnalysis

import warnings
warnings.filterwarnings('ignore')
```

```python
brupt=pd.read_csv('data.csv')

brupt.head()
```

```python
X=brupt.drop('Bankrupt?',axis=1)
y=brupt['Bankrupt?']
```

### Linear Descriminent Analysis

```python

lda=LinearDiscriminantAnalysis()
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params={'solver':['svd','lsqr','eigen']}

gcv=GridSearchCV(lda,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
```

### Quadratic Descriminent Analysis

```python
qda=QuadraticDiscriminantAnalysis()
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params={}

gcv=GridSearchCV(lda,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
```

# Digit dataset

```python
digitr=pd.read_csv('train.csv')

digitr.head()
```

```python
X=digitr.drop('label',axis=1)
y=digitr['label']
```

```python
### Linear Descriminent Analysis

lda=LinearDiscriminantAnalysis()
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params={'solver':['svd','lsqr','eigen']}

gcv=GridSearchCV(lda,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#Inferencing
bm_logistic=gcv.best_estimator_
test=pd.read_csv(r"test.csv")

y_pred=bm_logistic.predict(test)

submit_lg=pd.DataFrame({'ImageId':np.arange(1,28001),'label':y_pred})

submit_lg.to_csv(r"/home/dai/Desktop/Practical Machine Learning/Day-06/submitLDA.csv",index=False)

##0.86
```
