```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold,GridSearchCV,KFold
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.svm import SVR
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
```

```python
Z={'x1':[1,2,3],'x2':[3,9,4],'x3':[4,6,11],'y':['yes','no','no']}
df=pd.DataFrame(Z)
X=df.drop('y',axis=1)
y=df['y']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=23,stratify=y)
```

```python
dtc=DecisionTreeClassifier(random_state=23)
dtc.fit(X_train,y_train)
#Creating a graphical representation of Tree
plt.figure(figsize=(50,20))
plot_tree(dtc,feature_names=X_train.columns,class_names=['yes','no'], filled=True,fontsize=20)
plt.show()
```
