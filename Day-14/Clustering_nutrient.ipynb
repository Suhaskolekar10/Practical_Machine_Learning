```python
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
nutrient=pd.read_csv('nutrient.csv',index_col=0)
scalar=StandardScaler()
nutrientscaled=scalar.fit_transform(nutrient)
```

```python
#Calculate the linkage
mergings=linkage(nutrientscaled,method='average')
dendrogram(mergings,labels=list(nutrient.index),leaf_rotation=90,leaf_font_size=10)
plt.ylabel('Cluster Distance')
plt.show()
```

# SK-LEARN WITH hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.pipeline import Pipeline
clustering=AgglomerativeClustering(n_clusters=3,linkage="average")
pipe=Pipeline([('SCL',scalar),('CLUST',clustering)])
nutrientscaled=scalar.fit_transform(nutrient)
pipe.fit(nutrient)
print(clustering.labels_)
print(silhouette_score(nutrientscaled,clustering.labels_))
#print(silhouette_score(nutrient,clustering.labels_))
```
