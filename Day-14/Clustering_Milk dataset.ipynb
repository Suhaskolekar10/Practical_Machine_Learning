```python
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
milk=pd.read_csv('milk.csv',index_col=0)
scalar=StandardScaler()
milkscaled=scalar.fit_transform(milk)
```

```python
#Calculate the linkage
mergings=linkage(milkscaled,method='average')
dendrogram(mergings,labels=list(milk.index),leaf_rotation=90,leaf_font_size=10)
plt.ylabel('Cluster Distance')
plt.show()
```

```python
#Calculate the linkage
mergings=linkage(milkscaled,method='complete')
dendrogram(mergings,labels=list(milk.index),leaf_rotation=90,leaf_font_size=10)
plt.show()
```

```python
#Calculate the linkage
mergings=linkage(milkscaled,method='centroid')
dendrogram(mergings,labels=list(milk.index),leaf_rotation=90,leaf_font_size=10)
plt.show()
```

```python
#Calculate the linkage
mergings=linkage(milkscaled,method='single')
dendrogram(mergings,labels=list(milk.index),leaf_rotation=90,leaf_font_size=10)
plt.show()
```

# SK-LEARN WITH hierarcical clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.pipeline import Pipeline
clustering=AgglomerativeClustering(n_clusters=2,linkage="average")
pipe=Pipeline([('SCL',scalar),('CLUST',clustering)])
milkscaled=scalar.fit_transform(milk)
pipe.fit(milk)
print(clustering.labels_)
print(silhouette_score(milkscaled,clustering.labels_))
print(silhouette_score(milk,clustering.labels_))
```
