```python
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
US=pd.read_csv('USArrests.csv',index_col=0)
scalar=StandardScaler()
USscaled=scalar.fit_transform(US)
```

```python
#Calculate the linkage
mergings=linkage(USscaled,method='average')
dendrogram(mergings,labels=list(US.index),leaf_rotation=90,leaf_font_size=10)
plt.ylabel('Cluster Distance')
plt.show()
```

# SK-LEARN WITH hierarcical clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.pipeline import Pipeline
clustering=AgglomerativeClustering(n_clusters=2,linkage="average")
pipe=Pipeline([('SCL',scalar),('CLUST',clustering)])
pipe.fit(USscaled)
print(clustering.labels_)
print(silhouette_score(USscaled,clustering.labels_))
#print(silhouette_score(US,clustering.labels_))
```
