```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold,GridSearchCV,train_test_split,KFold
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import log_loss,accuracy_score,r2_score
from sklearn.ensemble import BaggingClassifier,BaggingRegressor
```

```python
concrete=pd.read_csv('Concrete_Data.csv')
X=concrete.drop('Strength',axis=1)
y=concrete[['Strength']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=23)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
lr_11=r2_score(y_test,y_pred)
print('Alone Linear=',lr_11)
```

```python
#########################  BAGGING   ######################################
bagg=BaggingRegressor(lr,n_estimators=15)
bagg.fit(X_train,y_train)
y_pred_probb=bagg.predict(X_test)
bag_11=r2_score(y_test,y_pred_probb)
print('Bagging=',bag_11)
```

```python
#With decisiontree
from sklearn.tree import DecisionTreeRegressor
dtc=DecisionTreeRegressor(random_state=23)
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
dtc_11=r2_score(y_test,y_pred)
print('Alone DTC=',dtc_11)
```

```python
#########################  BAGGING   ######################################
bagg=BaggingRegressor(dtc,n_estimators=15,random_state=23)
bagg.fit(X_train,y_train)
y_pred_probb=bagg.predict(X_test)
bag_11=r2_score(y_test,y_pred_probb)
print('Bagging=',bag_11)
```

# DECISION TREE

```python
concrete=pd.read_csv('Concrete_Data.csv')
X=concrete.drop('Strength',axis=1)
y=concrete['Strength']
```

```python
dtr=DecisionTreeRegressor(random_state=23)
bagg=BaggingRegressor(dtr,random_state=23)
print(bagg.get_params())
params={'estimator__max_depth':[None,3,5],'estimator__min_samples_split':[2,5,10],'estimator__min_samples_leaf':[1,5,10],'n_estimators':[10,30,50]}
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
gcv=GridSearchCV(bagg,param_grid=params,cv=kfold,verbose=3)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
